```python
#libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
filename='/content/drive/MyDrive/Global_import_of_cocoa_beans.xlsx'
df = pd.read_excel(filename)
df.head()
```

# Global import of cocoa beans in Americas

```python

df_americas = df['Region'] == 'Americas' 
axisX = df[df_americas].groupby(['Year']).nunique()
axisY = df[df_americas]['Import in tonnes'].values
year = np.array(axisX.index)
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(year, axisY, c= 'b', label='Americas', marker='+');
axes.set_xlim(left=2015,right=2020)
axes.set_xlabel('Year', fontsize = 16)
axes.set_ylabel('Import in Tonnes', fontsize = 16)
axes.set_title('Global import of cocoa beans in Americas (in Thousands)', fontsize = 16)
plt.grid()
axes.legend()
```

# Global import of cocoa beans in Europe

```python
df_europe = df['Region'] == 'Europe' 
axisX = df[df_europe].groupby(['Year']).nunique()
axisY = df[df_europe].groupby(['Import in tonnes']).nunique()
tonnes = np.array(axisY.index)
year = np.array(axisX.index)
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(year, tonnes, c= 'g', label='Europe',marker = '.');
axes.set_xlim(left=2015,right=2020)
axes.set_xlabel('Year', fontsize = 16)
axes.set_ylabel('Import in Tonnes', fontsize = 16)
axes.set_title('Global import of cocoa beans in Europe (in millions)', fontsize = 16)
plt.grid()
axes.legend()
```

# Global import of cocoa beans in Asia & Oceania

```python
df_Asia = df['Region'] == 'Asia & Oceania' 
axisX = df[df_Asia].groupby(['Year']).nunique()
axisY = df[df_Asia].groupby(['Import in tonnes']).nunique()
tonnes = np.array(axisY.index)
year = np.array(axisX.index)
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(year, tonnes, c= 'y', label='Asia & Oceania',marker = '*');
axes.set_xlim(left=2015,right=2020)
axes.set_xlabel('Year', fontsize = 16)
axes.set_ylabel('Import in Tonnes', fontsize = 16)
axes.set_title('Global import of cocoa beans in Asia & Oceania (in Thousands)', fontsize = 16)
plt.grid()
axes.legend()
```
